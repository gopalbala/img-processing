```python
book = cv2.imread('/content/page-book.jpg')
cv2_imshow(book)
```

```python
graybook = cv2.cvtColor(book, cv2.COLOR_BGR2GRAY)
cv2_imshow(graybook)
```

```python
val,thresh = cv2.threshold(graybook,127,255, cv2.THRESH_BINARY)
cv2_imshow(thresh)
```

```python
val, thresh = cv2.threshold(graybook,180,255, cv2.THRESH_BINARY)
cv2_imshow(thresh)
```

```python
val, thresh = cv2.threshold(graybook,140,255,cv2.THRESH_BINARY)
cv2_imshow(thresh)
```

**Otsu Method**

```python
val, otsu = cv2.threshold(graybook,0,255, cv2.THRESH_OTSU)
cv2_imshow(otsu)
```
